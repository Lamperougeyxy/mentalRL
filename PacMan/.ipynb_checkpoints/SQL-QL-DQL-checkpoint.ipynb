```python
def plotscore(name,avg_mean,avg_std,filename,title=None,legend=False):
    alpha1 = 1
    alpha2 = 0.2
    LINE_WIDTH = 2
    
    plt.rcParams['figure.figsize'] = [15,15]
    
    fig = plt.figure()
    ax1 = fig.add_subplot()
    colors = ['blue','red','black','green','purple','brown','pink','yellow','grey']
    cMap = ListedColormap(colors)
    
    for idx in np.arange(len(avg_mean)):
            ax1.plot(avg_mean[idx], alpha=alpha1, lw=LINE_WIDTH, label=name[idx],c=colors[idx])
#     ax1.plot(avg_mean[0], alpha=alpha1, lw=LINE_WIDTH, label=name[0],c='blue')
#     ax1.plot(avg_mean[1], alpha=alpha1, lw=LINE_WIDTH, label=name[1],c='red')
#     ax1.plot(avg_mean[2], alpha=alpha1, lw=LINE_WIDTH, label=name[2],c='black')
    if avg_std is not None:
        for idx in np.arange(len(avg_mean)):
            m=idx
            ax1.fill_between(np.arange(len(avg_mean[m]))+1, avg_mean[m] - avg_std[m]/np.sqrt(5), avg_mean[m] + avg_std[m]/np.sqrt(5), alpha=alpha2,color=colors[m])
#         m=0
#         ax1.fill_between(np.arange(len(avg_mean[m]))+1, avg_mean[m] - avg_std[m]/np.sqrt(5), avg_mean[m] + avg_std[m]/np.sqrt(5), alpha=alpha2,color='blue')
#         m=1
#         ax1.fill_between(np.arange(len(avg_mean[m]))+1, avg_mean[m] - avg_std[m]/np.sqrt(5), avg_mean[m] + avg_std[m]/np.sqrt(5), alpha=alpha2,color='red')
#         m=2
#         ax1.fill_between(np.arange(len(avg_mean[m]))+1, avg_mean[m] - avg_std[m]/np.sqrt(5), avg_mean[m] + avg_std[m]/np.sqrt(5), alpha=alpha2,color='black')
   
    ax1.set_xlabel('round')
    ax1.set_ylabel('avg score')
    if legend:
        ax1.legend()
    ax1.title.set_text(title)

#     ax1.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
    fig.savefig(filename, bbox_inches='tight')
```
